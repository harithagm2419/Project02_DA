```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'Purchase Amount (USD)' : 'purchase_amount'})
df.columns
```

```python
# creating the labels to with age group column

groups = ['Young Adult', 'Adult','Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels= groups)
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly' :14,
    'Weekly':7 , 
    'Monthly':30, 
    'Quarterly':90, 
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months': 90
}

# new column "purchase_frequency_days" to the no.of days 
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# printing together
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
# both the columns have same values ...
# checking whether all the rows are same
( df['discount_applied'] == df['promo_code_used']).all()
```

```python
# removing the unwanted column from the data ->  "promo_code_used"
df = df.drop('promo_code_used', axis=1)
```
